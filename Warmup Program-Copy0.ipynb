```python
import numpy as np
import scipy as sp
from scipy import integrate
from scipy import interpolate
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#automatic theta_o and print theta integral to sheet
theta_o = 0.0
phi = 0.0
Pdata = open("PT.csv", "a")
Gtheta = pd.read_csv('GT0.csv', sep = ',', header = None, names = ['theta','gamma'])
Ltheta = pd.read_csv('LT0.csv', sep = ',', header = None, names = ['theta','L'])
Btheta = np.sqrt(1-(1/((Gtheta['gamma'])**2)))

while (theta_o <= (41*np.pi/180)):
    print("observation angle (degrees) is: " + str(theta_o*180.0/np.pi))
#    Pdata.write("\n\ntheta o is: " + str(theta_o*180/np.pi))
    while (phi <= 2*np.pi):
        cosalph = (np.cos(theta_o)*np.cos(Ltheta['theta']*np.pi/180.0))+(np.sin(theta_o)*np.sin(Ltheta['theta']*np.pi/180.0)*np.cos(phi))
        num = Ltheta['L']*np.sin(Ltheta['theta']*np.pi/180.0)
        den = 4*np.pi*(Gtheta['gamma']**4)*((1-(Btheta*cosalph))**3)
        ans = (num/den)
    #    print(ans)
            #theta integral
        theta_int = sp.integrate.trapz(y = ans, x = Ltheta['theta'])
    #    theta_int = np.sum(ans)
        print("sum for phi (degrees) = " + str(phi*180.0/np.pi) + " is: " + str(theta_int))
        Pdata.write(str(phi*180.0/np.pi) + "," + str(theta_int) + "\n")
        phi = phi+(2*np.pi/36) #2*np.pi/n, count by 360/n degrees at a time
    print("\n")
    theta_o = theta_o+(1*np.pi/180.0) #n*np.pi/180, count by n degrees at a time
    phi = 0
Pdata.close()
```

```python
#split sheet into array for phi integral and compute
#will need to delete last row of printed csv from last part before using module

Ptheta = pd.read_csv('PT.csv', sep = ',', header = None, names = ['Phi','theta_int'])
L_o = open("L_obs.csv", "a")
n = int(input("n from theta_o calcualtion above is: "))
loop_no = 41
P = np.split(Ptheta,loop_no)
#P = np.split(Ptheta,41)
i = 0

while (i<= (loop_no-1)):
    integral = sp.integrate.trapz(y = P[i]['theta_int'], x = P[i]['Phi'])
    print("when observation angle is: " + str(n*i) + ", L observed is: " + str(integral))
    L_o.write(str(n*i) + "," + str(integral) + "\n")
    i=i+1
L_o.close()
```

```python
#L normalization
L_o = pd.read_csv('L_obs.csv', sep = ',', header = None, names = ['theta','L'])
L_o_Norm = open("L_obs_Norm.csv", "a")
Nval = max(L_o['L'])
i = 0
n = 1

while (i< loop_no):
    angle = str(n*i)
    Norm = (L_o['L'][i])/Nval
    print("when observation angle is: " + angle + ", normalized L observed is: " + str(Norm))
    L_o_Norm.write(angle + "," + str(Norm) + "\n")
    i=i+1
L_o_Norm.close()
```

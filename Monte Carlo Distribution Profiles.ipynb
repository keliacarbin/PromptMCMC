```python
## Core Luminosity Distribution log

import math as m
import numpy as np
import matplotlib.pyplot as plt
from random import randint
from random import random as randdec

Lmax = 1*(10**53)
Lstar = 2*(10**52)
Lmin = 5*(10**49)
Lalpha = 0.95
Lbeta = 2.0
plt.figure(figsize=(5,5), dpi=80)

def LumDistr(L):
    "generates the values of the lumoninosity distribution of sGRB's according to \
    eqn (3) in 1808.04831.pdf"
    if(Lmin <= L <= Lstar):
        ldistr = (L**(-Lalpha))/(Lstar**(-Lalpha))
    else:
        ldistr = (L**(-Lbeta))/(Lstar**(-Lbeta))
    return(ldistr)

#distribtion plot
loglum = np.linspace(Lmin, Lmax, 2000, endpoint=True)
luminosity =  np.array([])
for L in loglum:
    luminosityx = np.array([(L, LumDistr(L))])
    luminosity = np.append(luminosity,luminosityx)
LumXY = luminosity.reshape(2000,2)

#randomized luminosity generator and tester
Laccepted = np.array([])
while (len(Laccepted)<=10000):
    Nexp = randint(-2,3) + randdec()
    Lexp = randint(49,53) + randdec()
    N = 10**Nexp
    L = 10**Lexp
    if ((N <= LumDistr(L)) and (Lmin < L < Lmax)):
        Laccepted_n = np.array([L,N]) ## [luminosity in ergs maybe, arbitrary]
        Laccepted = np.append(Laccepted, Laccepted_n)
LAccepted = Laccepted.reshape(int(len(Laccepted)/2),2)

plt.xlabel('core luminosity')
plt.ylabel('N')
plt.ylim(5*10**-3,5*10**2)
x1 = LumXY[:,0]
y1 = LumXY[:,1]
plt.plot(x1,y1,label='distribution')
x2 = LAccepted[:,0]
y2 = LAccepted[:,1]
plt.scatter(x2,y2,label='gen\'d luminosites')
plt.xscale('log')
plt.yscale('log')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5), dpi=80)
plt.xlabel('luminosity (log)')

def plot_loghist(x, bins):
  hist, bins = np.histogram(x, bins=bins)
  logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
  plt.hist(x, bins=logbins, range=None, histtype='bar', edgecolor='black', ls='dashed', lw=1, fc=(0, 0, 0, 0.5))
  plt.xscale('log')

plot_loghist(LAccepted[:,0], 20)
```

```python
#Modded Gravitational Wave Detection

#redshift to distance
distance = c*RAccepted[:,0]/Hubble #Mpc
angleobs = TAccepted[:,0] #radians
dMax = 220 #Mpc

def comparison(d, theta):
    "makes the lhs of the comparison function (10) in 1808.04831 to test \
    validity of solutions generated"
    return (dMax/d) * np.sqrt((1 + (6*(np.cos(theta)**2)) + (np.cos(theta)**4)) / 8)

rgoodgws = 0
rgwsets = np.array([])
for dist in distance:
    for angle in angleobs:
        if (comparison(dist, angle) > 1.0):
            rgwsets_n = ([dist, angle]) ## [distance in Mpc, angle in radians]
            rgwsets = np.append(rgwsets, rgwsets_n)
            rgoodgws+=1
rGWSets = rgwsets.reshape(int(len(rgwsets)/2),2)

print(rGWSets)
print(f"number of valid combinations found is {rgoodgws}")
print(f"there were {len(TAccepted[:,0])*len(RAccepted[:,0])} possible")
print(f"accounting for luminosity independance, we can have {rgoodgws*len(LAccepted[:,0])}")
```

```python
#sGRB Detection from those who passed the modded gw test

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

angleobs = rGWSets[:,1] #radians
luminositiestest = LAccepted[:,0]

NormProfile = pd.read_csv('MCprofile.csv', sep = ',', header = None, names = ['theta','L'])

rgoodgams = 0
rgammasets = np.array([])
l = 0
i = 0

#cycling through core luminosities
while (l < len(luminositiestest)):
    #cycling through gravitational wave sets per luminosity
    while (i < len(angleobs)):
        #finding the closest angle obs to that one
        thetadjust = (NormProfile['theta']-rGWSets[i,1])**2
        thetmin = np.where(thetadjust == thetadjust.min())
        tmatch = max(thetmin[0])
        ithluminosity = NormProfile['L'][tmatch]
        #detectability comparison per gwset per luminsity
        reallum = ithluminosity*LAccepted[l,0]
        isolum = reallum/(4*np.pi*((rGWSets[i,0]*3.086*(10**24))**2))
        if (isolum >= (5.8*(10**(-7)))):
            rgammasets_n = ([rGWSets[i,0], rGWSets[i,1], LAccepted[l,0]]) ## [distance in Mpc, angle in radians, luminosity in ergs]
            rgammasets = np.append(rgammasets, rgammasets_n)
            rgoodgams+=1
        i+=1
    l+=1
    i=0

rGammaSets = rgammasets.reshape(int(len(rgammasets)/3),3)

print(rGammaSets)
print(f"number of valid gamma-ray detectable combinations found is {rgoodgams}")
```
